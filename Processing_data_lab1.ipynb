```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("dulieuxettuyendaihoc.csv")
data.head(10)
```

```python
X = data[len(data) - 10: ]
X = pd.DataFrame(X)
X.head(10)
```

```python
data.tail(10)
```

```python
data.fillna(0, inplace = True)
data.tail(10)
```

```python
data["T1"].fillna(data["T1"].mean(), inplace = False)
```

```python
def fillData(data, col): 
    return data[col].fillna(data[col].mean(), inplace = False)

for col in data.columns:
    datatype = str(data[col].dtype)
    if datatype in ["float64", "int64"]:
        fillData(data, col)
```

```python
'''
TBM = (T*2 + L + H + S + V*2 + X + D + N) / 10
'''
for i in ["1", "2", "3"]:
    data["TBM" + i] = np.round(((data["T" + i] * 2) + data["L" + i] + data["H" + i] + data["S" + i] + (data["V" + i] * 2) + data['X' + i] + data['D' + i] + data['N' + i]) / 10, 2)
```

```python
data.head(10)
```

# Bai 8

```python
def evaluate(score):
    if score < 5.0: return "Y"
    elif 5.0 <= score < 6.5: return "TB"
    elif 6.5 <= score < 8.0: return "K"
    elif 8.0 <= score < 9.0: return "G"
    return "XS"

for i in ["1", "2", "3"]:
    col_name = data["TBM" + i].values
    tmp = []
    for col in col_name: tmp.append(evaluate(col))
    data["XL" + i] = tmp
```

```python
data.head(10)
```

# Bai 9
# Normalization score by using Min Max Normalization

```python
def minMaxScaler(val_fl):
    return np.round((val_fl * 4) / 10, 2)


for i in ["1", "2", "3"]:
    col_name = data[f"TBM{i}"]
    np_arr = np.array(col_name)
    data[f"US_TBM{i}"] = minMaxScaler(np_arr)
```

```python
data.head(10)
```

# Bai 10

```python
'''
A = (DH1 * 2 + DH2 + DH3) / 4
B = (DH1 + DH2 * 2 + DH3) / 4
C = (DH1 + DH2 + DH3) / 3
'''
data_KT = data["KT"].values
dh1 = data["DH1"].values
dh2 = data["DH2"].values
dh3 = data["DH3"].values
value = []
for idx in range(len(data_KT)):
    if data_KT[idx] in ["A", "A1"]: value.append(1.0 if (dh1[idx] * 2 + dh2[idx] + dh3[idx]) / 4.0 >= 5.0 else 0.0)
    elif data_KT[idx] == "B": value.append(1.0 if (dh1[idx] + dh2[idx] * 2 + dh3[idx]) / 4.0 >= 5.0 else 0.0)
    else: value.append(1.0 if (dh1[idx] + dh2[idx] + dh3[idx]) / 3.0 >= 5.0 else 0.0)
data["KQXT"] = value
```
