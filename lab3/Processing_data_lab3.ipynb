```python
col = ["Id", "Name", "Age", "Weight", "m0006", "m0612", "m1218", "f0006", "f0612", "f1218"]
df = pd.read_csv("patient_heart_rate.csv", names = col)
df.head(5)
# Gán tên cột vào dataframe
```
