```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
df = pd.read_csv('titanic_disaster.csv')
df.head(10)
# đọc file csv bằng pandas và hiển thị 10 dòng đầu tiên
```

```python
seaborn.heatmap(df.isnull(), cmap = 'viridis')
plt.show()
# sử dụng biểu đồ heatmap (biểu đồ nhiệt trong seaborn) để visualize dữ liệu bị thiếu trong data.
```

```python
df[['FirstName', 'SecondName']] = df['Name'].str.split(',', expand = True)
df.drop(columns = ['Name'], inplace = True)
df.head(10)
# Tách cột Name thành 2 giá trị và sau đó gán chúng vào 2 cột mới là FirstName và SecondName. Sau đó dùng hàm drop để xóa cột Name
```

```python
df.replace({'male': 'M', 'female': 'F'}, inplace = True)
df.head(10)
# dùng hàm replace với tham số inplace = True có nghĩa là thay thế vào giá trị gốc bằng các giá trị mới.
```

```python
seaborn.boxplot(x = 'Pclass', y = 'Age', data = df)
plt.title('Box Plot Age')
plt.show()
# dùng hàm boxplot trong thư viện seaborn để visualize dữ liệu.
```

```python
avg_age = (int(df['Age'].mean()))
col_age = df['Age'].values
new_val = []
for age in col_age: new_val.append(avg_age if str(age) == 'nan' else int(age))
df['Age'] = new_val
seaborn.heatmap(df.isnull(), cmap = 'viridis')
plt.show()
# gán giá trị trung bình cho những chỗ bị null trong cột age.
# dùng biểu đồ heatmap để visualize dữ liệu bị thiếu trong data.
```

```python
new_val = []
col_age = df['Age'].values
for age in col_age:
    if age <= 12: new_val.append('Kid')
    elif 12 < age <= 18: new_val.append('Teen')
    elif 18 < age <= 60: new_val.append('Adult')
    elif age > 60: new_val.append('Older')
df['AgeGroup'] = new_val
df.head(10)
# Tạo cột mới là Age Group dựa trên các điều kiện.
```

```python
pref_name = []
col_name = df['SecondName'].values
for name in col_name:
    process_str = name.split('.')
    new_name = ' '.join(process_str[i] for i in range(1, len(process_str)))
    new_name.strip()
    pref_name.append(new_name)
df['SecondName'] = pref_name
df.head(10)
# Tách các từ như Mr, Master, ... ra khỏi second name.
```

```python
family_size = df['SibSp'] + df['Parch'] + 1
df['familySize'] = family_size
df.head(10)
# tính cột family size bằng công thức f = SibSP + Parch + 1
```

```python
def alone(num): return int(num == 0)
df['Alone'] = df['familySize'].apply(alone)
df.head(10)
# tính cột Alone dựa trên cột family size.
```

```python
def rename_cabin(name): return 'Unknown' if str(name) == 'nan' else str(name[0])
df['Cabin'] = df['Cabin'].apply(rename_cabin)
df.head(10)
# Thay thế lại các giá trị null trong cột bằng Unknown và chữ cái đầu tiên cho giá trị không null
```

```python
sex_survive = df.groupby(['Sex', 'Survived'])['Sex'].agg('count').unstack()
sex_survive.plot(kind = 'bar')
# trực quan dữ liệu của người sống sót và thiệt mạng trên từng giới tính bằng cách group by chúng lại.
# dùng hàm plot để visualize.
```

```python
df_survive = df[df['Survived'] == 1]
pc_class = df_survive.groupby(['Pclass'])['Pclass'].count()
plt.ylabel('count')
pc_class.plot.bar()
# tạo 1 data frame mới để lưu các cột có người sống sót (Survived = 1)
# sử dụng hàm plot để visualize dữ liệu.
```

```python
df_survive = df[df['Survived'] == 1]
sex_age_group = df_survive.groupby(['AgeGroup', 'Sex'])['AgeGroup'].count()
sex_age_group.plot.bar()
# tạo 1 data frame mới để lưu các cột có người sống sót (Survived = 1)
# sử dụng hàm plot để visualize dữ liệu.
```

```python
seaborn.catplot(x = 'familySize', y = 'Survived', data = df, kind = 'point')
plt.show()
# sử dụng hàm catplot để trực quan xác suất của dữ liệu dựa trên thông tin hành khách.
```

```python
seaborn.catplot(x = 'Fare', y = 'Survived', kind = 'point', data = df)
plt.show()
# sử dụng hàm catplot để trực quan xác suất của dữ liệu dựa trên thông tin vé
```

```python
seaborn.catplot(x = 'Pclass', y = 'Survived', hue = 'Sex', col = 'Embarked', data = df, kind = 'point')
plt.show()
# sử dụng hàm catplot để trực quan xác suất của dữ liệu dựa trên Pclass
```
